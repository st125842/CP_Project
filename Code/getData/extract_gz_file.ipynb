```python
import os
import gzip
import shutil

# Set the root directory to start the search
root_dir = 'DatasetPair/Task1/'  # Change this to your target directory

# Walk through all subdirectories
for foldername, subfolders, filenames in os.walk(root_dir):
    for filename in filenames:
        if filename.endswith('.gz'):
            gz_path = os.path.join(foldername, filename)
            output_path = os.path.join(foldername, filename[:-3])  # remove .gz extension

            # Extract the .gz file
            with gzip.open(gz_path, 'rb') as f_in:
                with open(output_path, 'wb') as f_out:
                    shutil.copyfileobj(f_in, f_out)

            print(f"Extracted: {gz_path} → {output_path}")
```
